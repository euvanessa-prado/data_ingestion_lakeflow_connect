###Parte 1: Criar o Catálogo

```python
%sql
CREATE CATALOG IF NOT EXISTS smart_claims_dev
COMMENT 'Catálogo principal para o projeto Smart Claims - Ambiente de Desenvolvimento'
```

```python
%sql
SHOW CATALOGS LIKE 'smart_claims_dev*'
```
